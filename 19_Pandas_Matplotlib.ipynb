```python
plt.figure(figsize=(16,4))
for i in range(50):
    plt.scatter(i,df.B[i],color="g",alpha=.2,s=.07*np.pi*df.B[i]**2)
plt.show()
```

```python
plt.figure(figsize=(16,4))
for i in range(50):
    plt.scatter(i,df.B[i],alpha=.2,s=.02*np.pi*df.B[i]**2)
plt.show()
```

```python
plt.figure(figsize=(16,4))
for i in range(50):
    if df.B[i]>150:
        plt.scatter(i,df.B[i],color="g",alpha=.4,s=.02*np.pi*df.B[i]**2,marker="*")
    else:
        plt.scatter(i,df.B[i],color="r",alpha=.1,s=.02*np.pi*df.B[i]**2)
plt.show()
```

```python
plt.figure(figsize=(16,4))
for i in range(50):
    if df.B[i]>150:
        plt.scatter(i,df.B[i],color="g",alpha=.4,s=.02*np.pi*df.B[i]**2,marker="*")
    else:
        plt.scatter(i,df.B[i],color="r",alpha=.1,s=.02*np.pi*df.B[i]**2,marker="^")
plt.show()
```

```python
plt.figure(figsize=(16,4))
for i in range(50):
    if df.B[i]>150:
        plt.scatter(i,df.B[i],color="g",alpha=.4,s=.02*np.pi*df.B[i]**2,marker="*")
        plt.text(i,df.B[i],df.B[i])
    else:
        plt.scatter(i,df.B[i],color="r",alpha=.1,s=.02*np.pi*df.B[i]**2,marker="^")

plt.show()
```

# Saving figure

```python
plt.figure(figsize=(16,4))
for i in range(50):
    if df.B[i]>150:
        plt.scatter(i,df.B[i],color="g",alpha=.4,s=.02*np.pi*df.B[i]**2,marker="*")
        plt.text(i,df.B[i],df.B[i])
    else:
        plt.scatter(i,df.B[i],color="r",alpha=.1,s=.02*np.pi*df.B[i]**2,marker="^")
plt.savefig("star_preformer.png")
plt.show()
```

# Controlling labels and ticks of graphs

```python
plt.figure(figsize=(16,4))
plt.bar(range(50),df.C)
plt.xticks(range(50))
plt.yticks(range(0,200,10))
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Bar-Graph")
plt.show()
```

```python
plt.figure(figsize=(16,4))
plt.bar(range(50),df.C)
plt.xticks(range(50),rotation=90)
plt.yticks(range(0,200,10))
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Bar-Graph")
plt.show()
```

```python
plt.figure(figsize=(16,4))
plt.bar(range(50),df.C)
plt.xticks(range(50),rotation=90)
plt.yticks(range(0,200,10))
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Bar-Graph")
plt.axis("off")
plt.show()
```

###### - subplotting

```python
d = np.random.randint(1,100,40).reshape(10,4)
df = pd.DataFrame(d,columns=list("ABCD"))
df
```

```python
plt.subplot(2,2,1)
plt.plot(range(10),df.A)
plt.subplot(2,2,2)
plt.scatter(range(10),df.B)
plt.subplot(2,2,3)
plt.bar(range(10),df.C)
plt.subplot(224)
plt.barh(range(10),df.D)
```

```python
f,a = plt.subplots(2,2)
a[0][0].plot(range(10),df.A)
a[0][1].scatter(range(10),df.B)
a[1][0].bar(range(10),df.C)
a[1][1].barh(range(10),df.D)
```
